## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
def playerpower(playerlist, pokedex, id):
    """Function accepts list of players, pokedex, and player id.  Returns total player power based on pokemon"""
    
    # Initialize totalpower variable
    totalpower = 0
    
    # Extract player_pokemon from specified arg player and convert to list: ownedpoke
    ownedpoke = list(playerlist[id]['player_pokemon'])

    # Iterate through ownedpoke list and searching through pokedex
    for i in ownedpoke:
        # Sum attack and defense stats of specified pokemon
        pokepower = pokedex[i]['attack'] + pokedex[i]['defense'] + pokedex[i]['special_attack'] 
        + pokedex[i]['special_defense']
        
        # Add pokepower to total
        totalpower += pokepower
    return totalpower

#Print output of function playerpower
print('The power for pokemon player '+ str(players[1]['player_name']) +' is: ' + str(playerpower(players, pokedex, 1)))
print('The power for pokemon player '+ str(players[2]['player_name']) +' is: ' + str(playerpower(players, pokedex, 2)))
````

## 9. Write a function to generate the full pokedex

---

Write a function that recreates the pokedex you made before, but with the data read in from the full pokemon file. The `PokedexNumber` should be used as the `pokemon_id` key values for the dictionary of pokemon.

Your function should:

1. Take the parsed pokedex information you created above as an argument.
2. Return a dictionary in the same format as your original pokedex you created before containing the information from the parsed full pokedex file.

To test the function, print out the pokemon with id = 100.

````python
def fillpokedex(pokedexinfo):
    """This function takes in a list of lists of pokemon information and generates a dictionary of dictionaries"""
    # split data into header and data. Also remove pokedex from header as used for key
    header = pokedexinfo[0]
    data = pokedexinfo[1:]
    
    # initialize final dictionary variable
    fullpoke = {}
    
    # for loop to iterate through data and build pokedex
    for row in data:
        
        # initialize variables
        pokefilled = {}
        statlist = []
        key = ""
        
        # for item in row list, set as key if index is 0, else append to statlist
        for i, col in enumerate(row):
            if i == 0:
                key = col
            else:
                statlist.append(col)
                
        # attach the statlist to the pokefilled dictionary
        for i, val in enumerate(statlist):
            pokefilled[header[i+1]] = statlist[i]
            
        #Assign the filled pokefilled to fullpoke with col (above) as key
        fullpoke[key] = pokefilled
    return fullpoke
    

            
full = fillpokedex(optipoke)
print(full)
````
